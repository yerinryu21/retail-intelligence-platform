```python
import pandas as pd

df = pd.read_csv('../data/processed/clean_retail.csv')

# Verify no cancelled orders remain
assert not df['InvoiceNo'].astype(str).str.startswith('C').any(), "Cancelled orders still present"
print("✅ No cancelled orders")

# Verify no missing CustomerIDs
assert df['CustomerID'].isnull().sum() == 0, "Missing CustomerIDs still present"
print("✅ No missing CustomerIDs")

# Verify no negative quantities
assert (df['Quantity'] > 0).all(), "Negative quantities still present"
print("✅ No negative quantities")

# Verify no negative prices
assert (df['UnitPrice'] > 0).all(), "Negative prices still present"
print("✅ No negative prices")

# Verify derived columns exist
for col in ['TotalPrice', 'Year', 'Month', 'DayOfWeek', 'Week', 'Date']:
    assert col in df.columns, f"Missing column: {col}"
print("✅ All derived columns present")

print(f"\n✅ Clean data verified: {df.shape[0]:,} rows, {df.shape[1]} columns")
print(df.head())
```
