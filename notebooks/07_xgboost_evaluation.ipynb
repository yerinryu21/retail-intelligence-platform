## XGBoost Evaluation — scale_pos_weight vs SMOTE


**Results:**

| Model | Precision | Recall | F1 | ROC-AUC | PR-AUC |
|------|----------:|-------:|---:|--------:|-------:|
| Logistic Regression | 0.54 | 0.81 | 0.65 | 0.80 | 0.620 |
| XGBoost (`scale_pos_weight`) | 0.52 | 0.76 | 0.61 | 0.78 | 0.578 |
| XGBoost (SMOTE) | 0.52 | 0.77 | 0.62 | 0.77 | 0.565 |



### Which imbalance strategy won?

`scale_pos_weight` narrowly beat SMOTE on PR-AUC (0.578 vs 0.565) and ROC-AUC (0.778 vs 0.773), while SMOTE had a marginally higher F1 and recall. The differences are small — neither strategy dominates decisively. Per the project's model-selection rule (highest PR-AUC), `scale_pos_weight` was selected and saved as the working churn model.

### Is XGBoost actually better than Logistic Regression here?

Not with default hyperparameters — both XGBoost variants scored lower PR-AUC than Day 2's Logistic Regression baseline (0.62). This wasn't the expected outcome going in, but it's a legitimate result worth taking seriously rather than dismissing.

**Possible reasons:**
- The dataset is relatively small (3,920 customers) with only 7 features — XGBoost's strength is capturing complex non-linear feature interactions, and there may not be enough complexity in this feature set for tree ensembles to have an edge over a well-specified linear model.

- The churn signal may be fairly linear in nature — features like Frequency and Monetary likely have a fairly direct, monotonic relationship with churn risk, which is exactly the kind of pattern Logistic Regression is built to capture  efficiently.

- Day 3's XGBoost models used fixed, untuned hyperparameters (`n_estimators=200, max_depth=4, learning_rate=0.1`). This is not necessarily XGBoost's ceiling — Day 4's hyperparameter tuning (RandomizedSearchCV) is the fairer test of whether XGBoost can outperform the linear baseline.
